```python
dataf = pd.read_csv("predictive_maintenance.csv", index_col ="UDI" )
dataf.drop(["Product ID", "Type"], axis = 1, inplace = True)

dataf
```

```python
dataf=pd.get_dummies(dataf)
dataf
```

```python

dataf.shape
```

```python
labels=np.array(dataf['Failure Type_Overstrain Failure'])
```

```python
dataf=dataf.drop('Failure Type_Overstrain Failure',axis=1)
```

```python
labels
```

```python
dataf
```

```python
dataf_list=list(dataf.columns)
dataf=np.array(dataf)
```

```python

dataf
```

```python

dataf_list
```

```python
from sklearn.model_selection import train_test_split
```

```python

train_dataf,test_dataf,train_labels,test_labels=train_test_split(dataf,labels,test_size=0.25,random_state=42)
```

```python
test_labels.shape
```

```python
test_dataf.shape
```

```python


from sklearn.ensemble import RandomForestRegressor
```

```python

rf=RandomForestRegressor(n_estimators=1000,random_state=42)
```

```python
rf.fit(train_dataf,train_labels) 
```

```python
predections=rf.predict(test_dataf)
```

```python
test_labels.shape
```
